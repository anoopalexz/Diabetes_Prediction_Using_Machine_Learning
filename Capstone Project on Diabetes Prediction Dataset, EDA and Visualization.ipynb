```python
# Two continous variable

plt.scatter(df['age'], df['bmi'])
plt.xlabel("age")
plt.ylabel("bmi")
plt.title("Scatter plot of age and bmi")
plt.savefig("Scatter plot of age and bmi.png")
plt.grid(True)
plt.show()
```
